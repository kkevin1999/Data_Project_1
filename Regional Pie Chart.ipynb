```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

data_file = Path("Resources/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time_-_ARCHIVED.csv")
data_df = pd.read_csv(data_file)
data_df
```

```python
data_df["submission_date"] = pd.to_datetime(data_df["submission_date"])
```

```python
data_df.dtypes
```

```python
latest_df = data_df.loc[data_df["submission_date"] == "2022-10-18"] 
latest_df
```

```python
cleaned_df = latest_df.groupby(["state", "tot_cases", "tot_death"])["submission_date"].max().reset_index()
cleaned_df
```

```python
West = ["WA", "OR", "CA", "NV", "ID", "MT", "WY", "UT", "CO", "AZ", "NM", "AK", "HI"]
Mid_West = ["ND", "SD", "NE", "KS", "MN", "IA", "MO", "WI", "IL", "IN", "MI", "OH"]
North_East = ["ME", "NH", "VT", "MA", "CT", "RMI", "NY", "PA", "NJ"]
South = ["TX", "OK", "AR", "LA", "MS", "TN", "KY", "AL", "FL", "GA", "SC", "NC", "VA", "WV", "DC", "MD", "DE"]

cleaned_df["Region"] = cleaned_df["state"].apply(lambda x: "West" if x in West else
                                                          ("Mid_West" if x in Mid_West else
                                                          ("North_East" if x in North_East else
                                                          ("South" if x in South else np.nan))))

cleaned_df
```

```python
cleaned_df = cleaned_df.dropna().reset_index(drop=True)

cleaned_df
```

```python
for i in cleaned_df:
    fatal_pct = (cleaned_df["tot_death"] / cleaned_df["tot_cases"]) * 100
    
cleaned_df["% of cases fatal"] = fatal_pct
cleaned_df
```

```python
West_data = cleaned_df.loc[(cleaned_df["Region"] == "West")]
Mid_West_data = cleaned_df.loc[(cleaned_df["Region"] == "Mid_West")]
South_data = cleaned_df.loc[(cleaned_df["Region"] == "South")]
North_East_data = cleaned_df.loc[(cleaned_df["Region"] == "North_East")]

NE_total = North_East_data["tot_cases"].sum()
MW_total = Mid_West_data["tot_cases"].sum()
South_total = South_data["tot_cases"].sum()
West_total = West_data["tot_cases"].sum()

sum_total = NE_total + MW_total + South_total + West_total

print(f"NE is {NE_total}, MW is {MW_total}, South is {South_total}, West is {West_total}")

print(f"The sum total is {sum_total}")
```

```python
values = [NE_total, MW_total, South_total, West_total]
labels = ["Northeast", "Midwest", "South", "West"]

plt.pie(values, labels = labels, autopct = "%1.1f%%")
plt.title("Total COVID-19 Cases Distributed by Region")

plt.savefig("output-3.png")

plt.show()
```
