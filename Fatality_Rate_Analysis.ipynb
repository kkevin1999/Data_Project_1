```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

data_file = Path("Resources/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time_-_ARCHIVED.csv")
data_df = pd.read_csv(data_file)
data_df
```

```python
data_df["submission_date"] = pd.to_datetime(data_df["submission_date"])
```

```python
data_df.dtypes
```

```python
latest_df = data_df.loc[data_df["submission_date"] == "2022-10-18"] 
latest_df
```

```python
cleaned_df = latest_df.groupby(["state", "tot_cases", "tot_death"])["submission_date"].max().reset_index()
cleaned_df
```

```python
for i in cleaned_df:
    fatal_pct = (cleaned_df["tot_death"] / cleaned_df["tot_cases"]) * 100
    
cleaned_df["% of cases fatal"] = fatal_pct
cleaned_df
```

```python
cleaned_df = cleaned_df.drop(3)
cleaned_df = cleaned_df.drop(11)
cleaned_df = cleaned_df.drop(13)
cleaned_df = cleaned_df.drop(28)
cleaned_df = cleaned_df.drop(39)
cleaned_df = cleaned_df.drop(44)
cleaned_df = cleaned_df.drop(45)
cleaned_df = cleaned_df.drop(47)
cleaned_df = cleaned_df.drop(54)

cleaned_df
```

```python
sorted_df = cleaned_df.sort_values(["% of cases fatal"], ascending = False).reset_index(drop = True)
sorted_df
```

```python
top_5 = sorted_df.head(5)
top_5
```

```python
state = top_5["state"]
fatality = top_5["% of cases fatal"]

plt.bar(state, fatality, alpha = 0.8)
plt.title("States with Highest Fatality Rates as Portion of Cases")
plt.xlabel("State")
plt.ylabel("% of COVID-19 cases fatal")

plt.show()
```

```python
bottom_5 = sorted_df.tail(5)
bottom_5
```

```python
state = bottom_5["state"]
fatality = bottom_5["% of cases fatal"]

plt.bar(state, fatality, alpha = 0.8)
plt.title("States with Lowest Fatality Rates as Portion of Cases")
plt.xlabel("State")
plt.ylabel("% of COVID-19 cases fatal")

plt.show()
```

```python
top_5_pops = [12972008, 6177957, 10912876, 7359197, 2113344]
top_5_medicare = [10216, 9994, 10274, 9614, 8343]
top_5_uninsured = [5.4, 9.3, 12.7, 10.6, 10.0]

top_5["Population (2022)"] = top_5_pops
top_5["Medicare Spending ($/yr)"] = top_5_medicare
top_5["% of Pop. Uninsured"] = top_5_uninsured

top_5
```

```python
top_meanpop = np.mean(top_5["Population (2022)"])
top_meanmed = np.mean(top_5["Medicare Spending ($/yr)"])
top_meanuninsured = np.mean(top_5["% of Pop. Uninsured"])

print(f"The mean population is {top_meanpop}.")
print(f"The mean spending on Medicare per year is ${top_meanmed}.")
print(f"The mean % of the state's population that are uninsured is %{top_meanuninsured}.")
```
